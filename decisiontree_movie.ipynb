```python
import pandas as pd

# Data
data = {
    'popcorn': [1,1,0,0,1,1,0],
    'soda':    [1,0,1,1,1,0,0],
    'age':     [7,12,18,35,38,50,83],
    'movie':   [0,0,1,1,1,0,0]
}

df = pd.DataFrame(data)
print(df)
```

```python
x = df[['popcorn', 'soda', 'age']]  # Features
y = df['movie']                     # Target
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='gini') 
model.fit(x, y)
```

```python
# Example prediction
new_data = [[1, 1, 25]]  # popcorn=1, soda=1, age=25
prediction = model.predict(new_data)

print("Prediction:", prediction)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plot_tree(model, feature_names=['popcorn','soda','age'], class_names=['No','Yes'], filled=True)
plt.show()
```
