```python
import pandas as pd
b=pd.read_csv("market.csv")
b
```

```python
head=b.columns[0].split(';')
head
```

```python
data=b.iloc[:,0].str.split(';',expand=True)
data
```

```python
data.columns=head
data
```

```python
data.isnull().sum()
```

```python
data=data.astype(int)
```

```python
data.shape
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
#pip install mlxtend
```

```python
fi=apriori(data,min_support=0.1, use_colnames=True)
print(fi)
```

```python
r=association_rules(fi,metric="confidence",min_threshold=0.3)
print(r)
```

```python
print(r[['antecedents','consequents','support','confidence','lift']].sort_values(by="lift",ascending=False))
```

```python
item_frequencies = data.sum().sort_values(ascending=False)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x=item_frequencies.head(10).values,
            y=item_frequencies.head(10).index)
plt.title('Top 10 Frequent Items')
plt.xlabel('Frequency')
plt.ylabel('Items')
plt.show()
```
