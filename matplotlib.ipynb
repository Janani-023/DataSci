```python
import pandas as pd
```

```python
d=pd.read_csv("StudentsPerformance.csv")
d
```

```python
d.columns
```

```python
import matplotlib.pyplot as plt
```

```python
d.groupby("gender")[["math score", 'reading score',
       "writing score"]].mean().plot(kind="bar")
plt.title("gender vs score")
plt.xlabel("gender")
plt.ylabel("score")
plt.legend()
plt.show()
```

```python
c=["5star","dairymilk","kitkat"]
s=[213,330,234]
plt.bar(c,s,color=["yellow","violet","red"])
plt.xlabel("chocolate companies")
plt.ylabel("sales in rs")
plt.show()
```

```python
marks=[89,93,45,67,92,97,54,53,52,100]
plt.hist(marks)
plt.show()
```

```python
plt.hist(d["math score"],bins=20)
plt.title("distribution of maths score")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show()
```

```python
#scatter plot
plt.scatter(d["math score"],d["reading score"])
plt.title("maths vs reading")
plt.xlabel("math score")
plt.ylabel("reading score")
plt.show()
```

```python
#line plot
avg_s=d[["math score",'reading score',"writing score"]].mean()
avg_s
```

```python
x=avg_s.index
y=avg_s.values
```

```python
plt.plot(x,y)
plt.title("maths vs reading")
plt.xlabel("math score")
plt.ylabel("reading score")
plt.show()
```
