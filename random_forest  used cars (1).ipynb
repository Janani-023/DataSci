```python
import pandas as pd
f=pd.read_csv("fullGas.csv")
f
```

```python
f.columns[f.isna().any()]
```

```python
f.isnull().sum()
```

```python
f=f.drop("Image_url",axis=1)
f
```

```python
f['car_age']=2026-f["Year"]
f['car_age']
```

```python
f["price_perkm"]=f["Price"]/(f["Mileage_km"]+1)
f["price_perkm"]
```

```python
f.info()
```

```python
fn=f.select_dtypes(include=["int64","float64"]).columns
fn
```

```python
f[fn]=f[fn].fillna(f[fn].median())
f[fn].isnull().sum()
```

```python
fo=f.select_dtypes(include="object").columns
for i in fo:
    f[i]=f[i].fillna(f[i].mode()[0])
    f[i]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
d={}
le=LabelEncoder()
for i in fo:
    f[i]=le.fit_transform(f[i])
    d[i]=le
```

```python
f.isnull().sum()
```

```python
x=f.drop("Condition",axis=1)
y=f["Condition"]
x.info()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42,class_weight="balanced")
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
